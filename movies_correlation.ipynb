```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8)

df=pd.read_csv(r'D:\movies.csv')
```

```python
# to check null values

null_values = df.isnull()

# Counting the number of null values in each column
null_counts = null_values.sum()

print("Null value counts in each column:")
print(null_counts)
```

```python

df_copy = df.copy()

# Check for non-finite values (NaN or inf) in the 'budget' column
nan_values = df_copy['budget'].isnull()
inf_values = df_copy['budget'].isin([np.inf, -np.inf])
non_finite_values = nan_values | inf_values
nan_values = df_copy['gross'].isnull()
inf_values = df_copy['gross'].isin([np.inf, -np.inf])
non_finite_values = nan_values | inf_values

if non_finite_values.any():
    print("Non-finite values found in the 'budget' column.")
    # Handle non-finite values as per your requirement, e.g., fillna, dropna, etc.

# Convert the 'budget' column to nullable integer data type (Int64)
df_copy['budget'] = df_copy['budget'].astype('Int64')
df_copy['gross'] = df_copy['gross'].astype('Int64')


# Check if conversion was successful
print(df_copy.dtypes)

# Now you can proceed with your analysis using the modified DataFrame 'df_copy'
```

```python
numeric_columns = df.select_dtypes(include='number')
correlation = numeric_columns.corr(method='pearson')
print(correlation)
```

```python
correlation = numeric_columns.corr(method='kendall')
sns.heatmap(correlation, annot=True)
plt.title("correlation heatmap")
plt.xlabel('features')

plt.show
```

```python
correlation = numeric_columns.corr(method='spearman')
sns.heatmap(correlation, annot=True)
plt.title("correlation heatmap")
plt.xlabel('features')

plt.show
```
